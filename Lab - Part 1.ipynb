```python
"""
encoding=utf8

1.Download data from: https://github.com/raficohen/medicalhebrew/archive/master.zip
2. Unzip
3.Assign workDir the path where you unzipped the data 
"""
import os
import urllib
import zipfile
urllib.urlretrieve("https://github.com/raficohen/medicalhebrew/archive/master.zip", filename="workshop.zip")
print "got file"
zipfile.ZipFile("workshop.zip").extractall(".")
print "extracted"
```

```python
"""
Change dir to the data folder
"""
try:
    os.chdir("medicalhebrew-master")
except:
    print "make sure we are in the correct folder"
    print "
filesDir = "sample"

"""
Sanity Check
"""
print os.listdir(filesDir)[0]

print open(os.path.join(filesDir,os.listdir(filesDir)[0])).read()
```

```python
"""
Read the texts in a python dictionary
"""
from codecs import open
docs = {}
for f in os.listdir(filesDir):
    fin = open(os.path.join(filesDir,f),"r","utf8")
    docs[f] = fin.read()
    
```

```python
"""
Sanity check!
"""
print docs.values()[300]
```

```python
"""
Dr. House question - can you tell the class? 
Hint - it's not lopus
"""
print docs.keys()[300].decode("cp1255")
```

```python
"""
Consturct our classifier
"""
def lungClass(text,rules):
    lung = False
    for word in rules:
        if word.decode("utf8") in text:
            lung = True
    return lung

rules = ["ריאה","גידול"]


for doc in docs:
    print lungClass(docs[doc],rules),doc, "lung" in doc  
```

```python
"""
Evaluate with numbers, calculate accuracy
"""

right = 0 
wrong = 0
for doc in docs:
    if lungClass(docs[doc],rules) == bool("lung" in doc):
        right +=1
    else:
        wrong +=1
print "right:",right,"wrong:", wrong, "accuracy: ",float(right)/(len(docs))
```

```python
"""
Precision / Recall and F1 Score
How are we doing on the lung cancer only?
"""

tp = 0
fp = 0
fn = 0
for doc in docs:
    lungLabel = bool("lung" in doc)#unicode("ריאה",encoding="utf8") in unicode(doc,encoding="cp1255")
    if lungClass(docs[doc],rules):
        if lungLabel:
            tp +=1
            
        else:
            fp +=1
    else:
        if lungLabel:
            fn +=1

precision = float(tp)/(fp+tp)
recall = float(tp)/(tp+fn)
f1 = 2*precision*recall/(precision+recall)
print "true positive:",tp,"false positive:",fp,"precision:",precision,"recall:",recall,"fscore:",f1
```
